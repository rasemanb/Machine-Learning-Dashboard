```python
import pandas as pd 
import json
from sqlalchemy import create_engine
import numpy as np
import datetime as datetime
```

```python
csv_file3 = "./movies_metadata.csv"
data3 = pd.read_csv(csv_file3)
data3
```

```python
data3['release_date'] = pd.to_datetime(data3['release_date'], errors='coerce').dt.year
```

```python
data3_df = data3[['id', 'title', 'genres', 'release_date', 'runtime', 'budget', 'revenue', 'vote_average', 'vote_count']]
data3_df
```

```python
data3_df.replace('..', np.nan, inplace=True)
drop_df = data3_df.dropna(how='any')
drop_df
```

```python
drop_df['genres'] = drop_df['genres'].str.replace("'", '"').apply(lambda x: [_["name"] for _ in json.loads(x)])
```

```python
variables = ['release_date', 'runtime', 'revenue', 'vote_count']
for column_name in variables:
 drop_df[column_name] = drop_df[column_name].astype(int)
```

```python
drop_df
```

```python
drop2_df = drop_df[drop_df['release_date'] > 1990]
drop2_df
```

```python
categories = drop2_df['genres'].str.join(sep='*').str.get_dummies(sep='*')
categories
```

```python
split_df = pd.DataFrame(drop2_df['genres'].tolist(), index=drop2_df.index)
split_df
```

```python
df2 = pd.concat([drop_df, categories], axis=1)
df2
```

```python
df2 = pd.concat([df2, split_df], axis=1)
df2
```

```python
df = pd.concat([drop2_df, split_df], axis=1)
df
```

```python
df = df.fillna('')
df
```

```python
df2 = df2.fillna('')
df2
```

```python
df2.to_csv('metadata_tableau_clean.csv')
```
